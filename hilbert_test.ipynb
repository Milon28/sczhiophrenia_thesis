```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, SubsetRandomSampler
import torchvision.transforms as transforms
from transformers import ViTForImageClassification, ViTConfig
import numpy as np
import pandas as pd
from PIL import Image
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

class EEGDataset(Dataset):
    def __init__(self, schizo_folder, healthy_folder, transform=None):
        """
        Custom dataset for EEG HHT plot images
        Args:
            schizo_folder: path to schizophrenia images
            healthy_folder: path to healthy images
            transform: image transformations
        """
        self.transform = transform
        self.images = []
        self.labels = []
        
        # Load schizophrenia images (label = 1)
        if os.path.exists(schizo_folder):
            for img_name in os.listdir(schizo_folder):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    self.images.append(os.path.join(schizo_folder, img_name))
                    self.labels.append(1)
        
        # Load healthy images (label = 0)
        if os.path.exists(healthy_folder):
            for img_name in os.listdir(healthy_folder):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    self.images.append(os.path.join(healthy_folder, img_name))
                    self.labels.append(0)
        
        print(f"Loaded {len([l for l in self.labels if l == 1])} schizophrenia images")
        print(f"Loaded {len([l for l in self.labels if l == 0])} healthy images")
        print(f"Total images: {len(self.images)}")
    
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        image_path = self.images[idx]
        label = self.labels[idx]
        
        # Load and convert image
        image = Image.open(image_path).convert('RGB')
        
        if self.transform:
            image = self.transform(image)
        
        return image, torch.tensor(label, dtype=torch.long)

class ViTClassifier(nn.Module):
    def __init__(self, num_classes=2, pretrained=True):
        """
        Vision Transformer for binary classification
        """
        super(ViTClassifier, self).__init__()
        if pretrained:
            self.vit = ViTForImageClassification.from_pretrained(
                'google/vit-base-patch16-224',
                num_labels=num_classes,
                ignore_mismatched_sizes=True
            )
        else:
            config = ViTConfig(
                image_size=224,
                patch_size=16,
                num_channels=3,
                num_labels=num_classes,
                hidden_size=768,
                num_hidden_layers=12,
                num_attention_heads=12,
                intermediate_size=3072
            )
            self.vit = ViTForImageClassification(config)
    
    def forward(self, x):
        outputs = self.vit(x)
        return outputs.logits

def train_model(model, train_loader, val_loader, epochs=50, lr=1e-4):
    """
    Train the ViT model
    """
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.01)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=5, factor=0.5)
    
    train_losses = []
    val_accuracies = []
    best_val_acc = 0.0
    best_model_state = None
    
    for epoch in range(epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        for batch_idx, (data, target) in enumerate(train_loader):
            data, target = data.to(device), target.to(device)
            
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
        
        # Validation phase
        model.eval()
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for data, target in val_loader:
                data, target = data.to(device), target.to(device)
                output = model(data)
                _, predicted = torch.max(output.data, 1)
                val_total += target.size(0)
                val_correct += (predicted == target).sum().item()
        
        val_acc = 100 * val_correct / val_total
        avg_train_loss = train_loss / len(train_loader)
        
        train_losses.append(avg_train_loss)
        val_accuracies.append(val_acc)
        
        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_state = model.state_dict().copy()
        
        scheduler.step(val_acc)
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.4f}, Val Acc: {val_acc:.2f}%')
    
    return best_model_state, train_losses, val_accuracies, best_val_acc

def evaluate_model(model, test_loader):
    """
    Evaluate the model and return predictions and true labels
    """
    model.eval()
    y_true = []
    y_pred = []
    
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            _, predicted = torch.max(output, 1)
            
            y_true.extend(target.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())
    
    return np.array(y_true), np.array(y_pred)

def plot_confusion_matrix(y_true, y_pred, fold_num, save_path=None):
    """
    Plot confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Healthy', 'Schizophrenia'],
                yticklabels=['Healthy', 'Schizophrenia'])
    plt.title(f'Confusion Matrix - Fold {fold_num}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    if save_path:
        plt.savefig(f'{save_path}/confusion_matrix_fold_{fold_num}.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return cm

def main():
    # Configuration
    SCHIZO_FOLDER = "D:/HHT/S"  # Update this path
    HEALTHY_FOLDER = "D:/HHT/H"       # Update this path
    BATCH_SIZE = 16
    EPOCHS = 50
    N_SPLITS = 5
    SAVE_DIR = "vit_eeg_results"
    
    # Create save directory
    os.makedirs(SAVE_DIR, exist_ok=True)
    
    # Image transformations
    transform_train = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(degrees=10),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    transform_val = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    # Load dataset
    full_dataset = EEGDataset(SCHIZO_FOLDER, HEALTHY_FOLDER, transform=transform_val)
    
    # Prepare for cross-validation
    labels = np.array(full_dataset.labels)
    skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)
    
    # Store results
    fold_results = []
    all_y_true = []
    all_y_pred = []
    
    print(f"\nStarting {N_SPLITS}-Fold Cross Validation")
    print("="*50)
    
    for fold, (train_idx, val_idx) in enumerate(skf.split(np.arange(len(full_dataset)), labels)):
        print(f"\nFold {fold + 1}/{N_SPLITS}")
        print("-" * 30)
        
        # Create data loaders for this fold
        train_dataset = EEGDataset(SCHIZO_FOLDER, HEALTHY_FOLDER, transform=transform_train)
        val_dataset = EEGDataset(SCHIZO_FOLDER, HEALTHY_FOLDER, transform=transform_val)
        
        train_sampler = SubsetRandomSampler(train_idx)
        val_sampler = SubsetRandomSampler(val_idx)
        
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, sampler=train_sampler)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, sampler=val_sampler)
        
        # Initialize model
        model = ViTClassifier(num_classes=2, pretrained=True).to(device)
        
        # Train model
        print(f"Training fold {fold + 1}...")
        best_model_state, train_losses, val_accuracies, best_val_acc = train_model(
            model, train_loader, val_loader, epochs=EPOCHS
        )
        
        # Load best model and evaluate
        model.load_state_dict(best_model_state)
        y_true_fold, y_pred_fold = evaluate_model(model, val_loader)
        
        # Calculate metrics
        fold_accuracy = accuracy_score(y_true_fold, y_pred_fold)
        
        # Plot confusion matrix
        cm = plot_confusion_matrix(y_true_fold, y_pred_fold, fold + 1, SAVE_DIR)
        
        # Store results
        fold_results.append({
            'fold': fold + 1,
            'accuracy': fold_accuracy,
            'best_val_acc': best_val_acc,
            'confusion_matrix': cm
        })
        
        all_y_true.extend(y_true_fold)
        all_y_pred.extend(y_pred_fold)
        
        # Save best model for this fold
        torch.save({
            'fold': fold + 1,
            'model_state_dict': best_model_state,
            'accuracy': fold_accuracy,
            'train_losses': train_losses,
            'val_accuracies': val_accuracies
        }, f'{SAVE_DIR}/best_model_fold_{fold + 1}.pth')
        
        print(f"Fold {fold + 1} Results:")
        print(f"Accuracy: {fold_accuracy:.4f}")
        print(f"Best Validation Accuracy: {best_val_acc:.2f}%")
        
        # Classification report for this fold
        print("\nClassification Report:")
        print(classification_report(y_true_fold, y_pred_fold, 
                                  target_names=['Healthy', 'Schizophrenia']))
    
    # Overall results
    print("\n" + "="*50)
    print("OVERALL CROSS-VALIDATION RESULTS")
    print("="*50)
    
    # Calculate mean and std of accuracies
    accuracies = [result['accuracy'] for result in fold_results]
    mean_accuracy = np.mean(accuracies)
    std_accuracy = np.std(accuracies)
    
    print(f"Mean Accuracy: {mean_accuracy:.4f} ± {std_accuracy:.4f}")
    print(f"Individual Fold Accuracies: {[f'{acc:.4f}' for acc in accuracies]}")
    
    # Overall confusion matrix
    overall_cm = confusion_matrix(all_y_true, all_y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(overall_cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Healthy', 'Schizophrenia'],
                yticklabels=['Healthy', 'Schizophrenia'])
    plt.title('Overall Confusion Matrix (All Folds)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.savefig(f'{SAVE_DIR}/overall_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Overall classification report
    print("\nOverall Classification Report:")
    print(classification_report(all_y_true, all_y_pred, 
                              target_names=['Healthy', 'Schizophrenia']))
    
    # Save results summary
    results_df = pd.DataFrame(fold_results)
    results_df.to_csv(f'{SAVE_DIR}/cross_validation_results.csv', index=False)
    
    # Find and save the best overall model
    best_fold_idx = np.argmax(accuracies)
    best_fold = fold_results[best_fold_idx]
    
    print(f"\nBest performing fold: {best_fold['fold']} with accuracy: {best_fold['accuracy']:.4f}")
    
    # Save summary
    summary = {
        'mean_accuracy': mean_accuracy,
        'std_accuracy': std_accuracy,
        'best_fold': best_fold['fold'],
        'best_accuracy': best_fold['accuracy'],
        'overall_confusion_matrix': overall_cm.tolist()
    }
    
    import json
    with open(f'{SAVE_DIR}/summary.json', 'w') as f:
        json.dump(summary, f, indent=2)
    
    print(f"\nAll results saved in: {SAVE_DIR}/")
    print("Files saved:")
    print("- Individual fold models: best_model_fold_*.pth")
    print("- Confusion matrices: confusion_matrix_fold_*.png")
    print("- Overall confusion matrix: overall_confusion_matrix.png")
    print("- Results summary: cross_validation_results.csv")
    print("- Summary statistics: summary.json")

if __name__ == "__main__":
    main()
```
